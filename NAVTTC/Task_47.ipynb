## Pandas - Cleaning Data

```python
import pandas as pd

df = pd.read_csv('AB_NYC_2019.csv')

new_df = df.dropna()

print(new_df.to_string())
```

```python
df.dropna(inplace = True)

print(df.to_string())
df.head()
```

```python
df.fillna(130, inplace = True)
```

```python

df.fillna({"latitude": 130}, inplace=True)
```

```python

x = df["latitude"].mean()

df.fillna({"": x}, inplace=True)
```

```python

x = df["latitude"].median()

df.fillna({"latitude": x}, inplace=True)
```

```python
df["latitude"].mode()[0]

df.fillna({"latitude": x}, inplace=True)
```

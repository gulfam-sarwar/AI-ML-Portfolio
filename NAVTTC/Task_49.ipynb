## Machine Learning - Mean Median Mode

```python
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)

```

```python

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

```python
x = numpy.median(speed)

print(x)
```

```python
from scipy import stats
x = stats.mode(speed)

print(x)
```

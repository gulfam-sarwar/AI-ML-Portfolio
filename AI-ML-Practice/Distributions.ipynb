Use when:
Data naturally average ke around hota hai.

Examples:
Example 1: Students ke exam marks
Example 2: Human heights

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t

# Style
sns.set(style="whitegrid")
np.random.seed(42)

# -----------------------------
# 1) NORMAL DISTRIBUTION
# -----------------------------
# Example 1: Students Marks
normal_marks = np.random.normal(loc=70, scale=10, size=200)

# Example 2: Heights
normal_heights = np.random.normal(loc=170, scale=7, size=200)

plt.figure(figsize=(8,5))
sns.histplot(normal_marks, bins=20, kde=True)
plt.title("Normal Distribution - Students Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(normal_heights, bins=20, kde=True)
plt.title("Normal Distribution - Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()

```

BERNOULLI 
ek trial ko represent karti hai jisme 2 outcomes hotay hain:
Success (1) ya Failure (0).

```python
# Style
sns.set(style="whitegrid")
np.random.seed(42)

# -----------------------------
# BERNOULLI DISTRIBUTION
# -----------------------------
# Example 1: Coin Toss
# p = probability of head = 0.5
bernoulli_coin = np.random.binomial(n=1, p=0.5, size=200)

# Example 2: Student Pass/Fail
# p = probability of pass = 0.7
bernoulli_pass = np.random.binomial(n=1, p=0.7, size=200)

# Plot 1
plt.figure(figsize=(8,5))
sns.countplot(x=bernoulli_coin)
plt.title("Bernoulli Distribution - Coin Toss")
plt.xlabel("Outcome (0 = Tail, 1 = Head)")
plt.ylabel("Count")
plt.show()

# Plot 2
plt.figure(figsize=(8,5))
sns.countplot(x=bernoulli_pass)
plt.title("Bernoulli Distribution - Student Pass/Fail")
plt.xlabel("Outcome (0 = Fail, 1 = Pass)")
plt.ylabel("Count")
plt.show()
```

BINOMIAL
Fixed number of trials ho aur result 2 hi ho:

Success / Failure
Pass / Fail
Head / Tail

```python
# 3) BINOMIAL DISTRIBUTION
# -----------------------------
# Example 1: 10 coin tosses
binomial_coin = np.random.binomial(n=10, p=0.5, size=200)
# Example 2: 20 students pass/fail
binomial_pass = np.random.binomial(n=20, p=0.7, size=200)

plt.figure(figsize=(8,5))
sns.histplot(binomial_coin, bins=11, kde=False)
plt.title("Binomial Distribution - Heads in 10 Coin Tosses")
plt.xlabel("Number of Heads")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(binomial_pass, bins=15, kde=False)
plt.title("Binomial Distribution - Pass Students out of 20")
plt.xlabel("Number of Passed Students")
plt.ylabel("Frequency")
plt.show()
```

POISSON

Kisi fixed time/area mein events count karne ho.

Examples:
Example 1: 1 hour mein customer calls
Example 2: 1 day mein website errors

```python
# 4) POISSON DISTRIBUTION
# -----------------------------
# Example 1: Calls per hour
poisson_calls = np.random.poisson(lam=4, size=200)

# Example 2: Website errors per day
poisson_errors = np.random.poisson(lam=2, size=200)

plt.figure(figsize=(8,5))
sns.histplot(poisson_calls, bins=10, kde=False)
plt.title("Poisson Distribution - Calls per Hour")
plt.xlabel("Number of Calls")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(poisson_errors, bins=10, kde=False)
plt.title("Poisson Distribution - Website Errors per Day")
plt.xlabel("Number of Errors")
plt.ylabel("Frequency")
plt.show()
```

Time between events measure karna ho.

Examples:
Example 1: 2 customers ke aane ke darmiyan time
Example 2: Machine failure tak ka waiting time

```python
# 5) EXPONENTIAL DISTRIBUTION
# -----------------------------
# Example 1: Time between customers (avg 5 min)
exp_customers = np.random.exponential(scale=5, size=200)

# Example 2: Machine failure waiting time (avg 10 hours)
exp_failure = np.random.exponential(scale=10, size=200)

plt.figure(figsize=(8,5))
sns.histplot(exp_customers, bins=20, kde=True)
plt.title("Exponential Distribution - Time Between Customers")
plt.xlabel("Time (minutes)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(exp_failure, bins=20, kde=True)
plt.title("Exponential Distribution - Machine Failure Waiting Time")
plt.xlabel("Time (hours)")
plt.ylabel("Frequency")
plt.show()
```

T-Distribution


Use when:
Sample small ho
Population standard deviation unknown ho
Statistical testing mein use hoti hai
Examples:
Example 1: 10 students ke marks ka analysis
Example 2: 12 employees ke working hours

```python
# 6) T-DISTRIBUTION
# -----------------------------
# Example 1: Small sample marks
t_marks = t.rvs(df=10, size=200) * 10 + 70

# Example 2: Small sample working hours
t_hours = t.rvs(df=8, size=200) * 2 + 8

plt.figure(figsize=(8,5))
sns.histplot(t_marks, bins=20, kde=True)
plt.title("T-Distribution - Small Sample Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(t_hours, bins=20, kde=True)
plt.title("T-Distribution - Small Sample Working Hours")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```
